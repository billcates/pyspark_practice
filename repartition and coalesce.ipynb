####Repartition and Coalesce

```
df1=spark.read.format("csv").option("header","true").load("dbfs:/FileStore/shared_uploads/billcates.m@lntinfotech.com/sachin_batting_scores.xls")
```

```
#print the number of partitions
df1.rdd.getNumPartitions()
```

By default spark creates 1 partiton for every 128mb of the file.

when Spark reads a compressed file, it will create a single partition for the data, which will contain all the records in the file. This means that all the processing of the data in the file will be done by a single task on a single executor core, which may result in poor performance if the file is very large

```
display(df1.limit(10))
```

####Repartition

```
%fs
mkdirs /user/practice/sachin
```

```
df1.repartition(4).write.mode("overwrite").partitionBy("versus").parquet("/user/practice/sachin")
```

```
%fs
ls /user/practice/sachin/versus=Australia/
```

####Coalesce

```
df1.coalesce(2).write.mode("overwrite").partitionBy("versus").parquet("/user/practice/sachin")
```
