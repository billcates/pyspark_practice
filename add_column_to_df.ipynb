```
df=spark.read.format("csv").options(header='true', inferSchema='true', delimiter=',') .load("dbfs:/FileStore/shared_uploads/billcates.m@lntinfotech.com/sachin_batting_scores.xls")
```

```
df.printSchema()
```

```
display(df.limit(5))
```

```
import pyspark.sql.functions as fn
```

###when statement

```
df1=df.withColumn("better_strike_rate",fn.when(fn.col("strike_rate")>100,"good").otherwise("no"))
```

```
display(df1.filter(df1.better_strike_rate=="good").limit(10))
```

##using udf functions

```
from pyspark.sql.types import StringType
# @udf(returnType=StringType())
# def my_udf(col):
#     # Custom transformation logic
#     return transformed_value

#we can use decorators as wel



def my_fn(col):
    if col>100:
        return "good"
    else:
        return "not bad"
my_udf=fn.udf(my_fn,StringType())
```

```
df3=df.withColumn("better_strike_rate",my_udf(fn.col("strike_rate")))
```

```
display(df3.limit(5))
```
