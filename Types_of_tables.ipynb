##Managed and Unmanaged table(External table)

###Managed table
spark stores both,metadata and data available

###External table
spark stores only the metadata and controls the data
while dropping the table it removes the metadata only, but data would be there in the location

```
df = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/billcates.m@lntinfotech.com/circuits.csv")
```

```
df.write.saveAsTable("circuits_csv")
```

```
df.rdd.getNumPartitions()
```

```
%sql
describe table extended circuits_csv
```

```
%sql
select * from circuits_csv
```

```
%sql
drop table circuits_csv
```
