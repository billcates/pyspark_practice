```
%sql
create table m_change_feed
(name STRING,id Int,age int)
using delta
tblproperties (delta.enableChangeDataFeed= true)
```

```
%sql
insert into m_change_feed
values("billy",1,23),("prakriti",2,19)
```

```
%sql
select * from table_changes("m_change_feed",1)
```

```
%sql
update m_change_feed
set name="sam"
where name='prakriti'
```

```
%sql
select * from table_changes("m_change_feed",1)
```

```
%sql
delete from m_change_feed
where name='billy'
```

```
%sql
select * from m_change_feed
```

```
%sql
select * from table_changes("m_change_feed",3)
```

```
spark.sql('insert into m_change_feed values("bill",3,21)')
```

```
%sql
select * from table_changes("m_change_feed",1,4) order by _commit_timestamp
```

Conver the table data to df so that we can query the data

```
df=spark.sql('select * from table_changes("m_change_feed",1,4) order by _commit_timestamp')
```
