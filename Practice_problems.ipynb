Merge two dataframes

```
data=[(1,2,3,4),(5,6,7,8)]
columns=['a','b','c','d']
df=spark.createDataFrame(data=data,schema=columns)
```

```
data=[(9,10,11,12),(13,1,17,18)]
columns=['a','b','c','d']
df1=spark.createDataFrame(data=data,schema=columns)
```

```
df=df.union(df1)
```

```
display(df)
```

Explode Columns

```
data=[(1,[1,2,3]),(5,[6,7,8])]
columns=['a','b']
df=spark.createDataFrame(data=data,schema=columns)
```

```
display(df)
```

```
from pyspark.sql.functions import explode
import pyspark.sql.functions as f
```

```
final_df=df.select("a",f.explode("b").alias("names"))
```

```
display(final_df)
```

Regular Expression

```
df = spark.createDataFrame([["Alex", 20], ["Bob", 30],["null",40]], ["name", "age"])
df.show()
```

```
df=df.filter(f.col("name").rlike('^B'))
```

```
display(df)
```

```
df1=df.select(f.count(f.when(f.col("name"),isNull())))
```

```
display(df1)
```

handle multi delimiters

```
df = spark.createDataFrame([["Alex", 20,"7|8|9"], ["Bob", 30,"2|3|4"]], ["name", "age","marks"])
df.show()
```

we have delimiter pipe

```
df1=df.withColumn("marks",f.split("marks",'\\|'))
```

```
display(df1)
```

```
df2=df.withColumn("physics",f.split("marks",'\\|')[0])
```

```
display(df2)
```

```
df2=df.withColumn("chem",f.split("marks",'\\|')[2]).withColumn("math",f.split("marks",'\\|')[1])
```

```
df2=df2.drop("marks")
```

```
display(df2)
```
