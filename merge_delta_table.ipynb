```
df = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/billcates.m@lntinfotech.com/circuits.csv")
```

```
import pyspark.sql.functions as fn
df=df.groupBy('name').agg(fn.max("circuitId").alias("id"))
display(df)
```

```
#write to a delta table and save as delta file
df.write.format("delta").option("overwriteSchema", "true").mode("overwrite").saveAsTable("play_df")
```

```
display(df.limit(5))
```

```
%sql
create table circuits_df3
using delta
as select name,id from play_df
```

```
%sql
select max(id) from play_df --limit 10
```

```
%sql
insert into play_df
values("billy",45)
```

```
%sql
merge into circuits_df3
using play_df
on circuits_df3.name=play_df.name
when matched then update  set *
when not matched then insert *
```
