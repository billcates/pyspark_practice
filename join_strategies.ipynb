Brroadcast join

```
data=[(x,) for x in range(10)]
```

```
columns=['id']
```

```
df=spark.createDataFrame(data,schema=columns)
```

```
display(df)
```

```
data1=[(x,) for x in range(5,7)]
df1=spark.createDataFrame(data1,schema=columns)
```

If one table is smaller than other and it is set in config then automatically broadcast join takes place
but manually we can specify as df.join(broadcast(df1),id)

```
df2=df.join(df1,df.id==df1.id)
```

```
display(df2)
```

SHuffle hash join

First shuffles and moves data into executors such that same value of join key in same executor
and then hash join

both shuffle hash join and broadcast join will work only for equi join

Shuffle Sort Merge Join

Shuffles and then moves data based on join keys to the executors and it sorts and then it joins
