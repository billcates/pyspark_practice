How to read csv from nth line

```
#df1 = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/billcates.m@lntinfotech.com/pageview.csv")
rdd=spark.sparkContext.textFile('dbfs:/FileStore/shared_uploads/billcates.m@lntinfotech.com/pageview.csv')
```

```
filtered_rdd=rdd.zipWithIndex().filter(lambda  x:x[1]>=8).keys()
```

```
rdd_f=filtered_rdd.map(lambda l:l.split(','))
```

```
df=spark.createDataFrame(rdd_f)
```

```
display(df)
```

```
import pyspark.sql.functions as f
df=df.groupBy("_2").agg(f.sum("_2"),f.avg("_3"))
```
