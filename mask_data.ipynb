```
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
```

```
def mask(s):
    tmp=s.split('@')[0]
    l=len(tmp)
    res=tmp[0]+('X'*(l-2))+s[l-1:]
    return res
```

```
mask=udf(mask,StringType())
```

```
df3=df1.withColumn('email',mask(df1['email']))
```
